```python
from sqlalchemy import create_engine
import pandas as pd

dataCSV = pd.read_csv('train.csv')

engine = create_engine('sqlite:///:memory:')

dataCSV.to_sql('data_table', engine)

res1 = pd.read_sql_query('SELECT * FROM data_table', engine)
print('Result 1')
print(res1)
print('')
```

```python
res2 = pd.read_sql_query('SELECT Embarked,sum(Fare) FROM data_table group by Embarked', engine)
print('Result 2')
print(res2)
```

```python
res2 = pd.read_sql_query('SELECT count(Name) FROM data_table WHERE Sex = (?)', engine, params = ['female'])
print('Result 2')
print(res2)
```

```python
res2 = pd.read_sql_query('SELECT count(Name) FROM data_table WHERE Sex = (?)', engine, params = ['male'])
print('Result 2')
print(res2)
```

```python
res2 = pd.read_sql_query('SELECT count(Name) FROM data_table WHERE Embarked = (?)', engine, params = ['C'])
print('Result 2')
print(res2)
```

```python
res2 = pd.read_sql_query('SELECT count(Name) FROM data_table WHERE Embarked = (?)', engine, params = ['S'])
print('Result 2')
print(res2)
```
